## 1. Matplotlib vs Seaborn - Side-by-Side Plot Comparison

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Line Plot:

```python
# Matplotlib
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
plt.plot(x, y)
plt.title("Matplotlib Line Plot")
plt.show()

# Seaborn
import pandas as pd
df = pd.DataFrame({"x": x, "y": y})
sns.lineplot(data=df, x="x", y="y")
plt.title("Seaborn Line Plot")
plt.show()
```

### Histogram:

```python
# Matplotlib
plt.hist(df["y"])
plt.title("Matplotlib Histogram")
plt.show()

# Seaborn
sns.histplot(df["y"])
plt.title("Seaborn Histogram")
plt.show()
```

### Multiple Subplots:

```python
# Matplotlib
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, y)
ax1.set_title("Plot 1")
ax2.plot(x, [i**3 for i in x])
ax2.set_title("Plot 2")
plt.tight_layout()
plt.show()

# Seaborn + FacetGrid
df["y2"] = [i**3 for i in x]
long_df = pd.melt(df, id_vars=["x"], value_vars=["y", "y2"], var_name="Type", value_name="Value")
sns.relplot(data=long_df, x="x", y="Value", col="Type", kind="line")
plt.show()
```

### Styled Line:

```python
# Matplotlib
plt.plot(x, y, linestyle="--", color="red", marker="o")
plt.title("Styled Line")
plt.show()

# Seaborn
sns.lineplot(x=x, y=y, linestyle="--", color="red", marker="o")
plt.title("Styled Seaborn Line")
plt.show()
```
