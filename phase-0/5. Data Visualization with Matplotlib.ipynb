# 1. Data Visualization with Matplotlib

> **Story:** Imagine trying to explain a story just with numbers - hard, right? But show it with a picture, and it instantly makes sense. That’s the magic of data visualization.

###  Setup:

```python
import matplotlib.pyplot as plt
```

### Scatter Plot

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 8, 7]

plt.scatter(x, y)
plt.title("Simple Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### Histogram

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5, 5]
plt.hist(data, bins=5)
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

###  Interpreting Histograms

- Look for **skewness**: is the data left-heavy or right-heavy?
- Check for **spread**: how wide is the distribution?
- Identify **peaks**: where most values are concentrated

## 2. Customizing Visualizations with Matplotlib

> **Goal:** Learn how to add style and structure to your plots so they’re clear and beautiful.

### Objectives

Students should be able to:

- Create a line plot
- Plot multiple graphs on the same axes
- Customize axis limits and ticks
- Change line styles and colors
- Understand figure vs. axes
- Create multiple subplots

### Line Plot

```python
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

### Multiple Lines

```python
x = [0, 1, 2, 3, 4]
plt.plot(x, [i**2 for i in x], label="x^2")
plt.plot(x, [i**3 for i in x], label="x^3")
plt.legend()
plt.title("Multiple Lines")
plt.show()
```

### Axis Limits and Ticks

```python
plt.plot(x, y)
plt.xlim(0, 5)
plt.ylim(0, 20)
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 5, 10, 15])
plt.title("Custom Axes")
plt.show()
```

### Styles and Colors

```python
plt.plot(x, y, linestyle="--", color="red", marker="o")
plt.title("Styled Line")
plt.show()
```

###  Figure vs. Axes

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Using Axes")
ax.set_xlabel("X")
ax.set_ylabel("Y")
```

- `fig` is the whole canvas.
- `ax` is the area where data is drawn.

### Subplots

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, y)
ax1.set_title("Plot 1")
ax2.plot(x, [i**3 for i in x])
ax2.set_title("Plot 2")
plt.tight_layout()
plt.show()
```
