## Common Mistakes in Data Visualization (with Code Examples)

### Don't: Use a Pie Chart (most of the time)

**Why?**

- Humans are not great at comparing angles.
- Small differences are hard to spot.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Pie Chart Example (Not Recommended)

sizes = [40, 30, 30]
labels = ["A", "B", "C"]
plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Pie Chart")
plt.show()
```

###  Don't: Get Your Percentages Wrong

```python
# Incorrect percentages - adds up to 110%
sizes = [50, 30, 30]  # 50 + 30 + 30 = 110%
labels = ["X", "Y", "Z"]
plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Wrong Percentages")
plt.show()
```

### Don't: Make Your Visualizations Too Busy

```python
# Overloaded plot

x = np.linspace(0, 10, 100)
for i in range(1, 6):
    plt.plot(x, np.sin(x + i), label=f"Line {i}", linestyle="--", marker="o")
plt.legend()
plt.grid(True)
plt.title("Too Many Lines")
plt.show()
```

###  Don't: Use 3D Charts

```python
# 3D Plot (Can distort perception)
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.array(([1,2,3,1,2,3]))
y = np.array([4,5,6,5,6,7])
z = np.array([7,8,9,6,5,4])

dx = 0.3
dy = 0.3
dz = z

colors = ['b', 'g', 'r', 'c', 'm', 'y']

for i in range(len(z)):
    ax.bar3d(x[i], y[i], 0, dx, dy, dz, color=colors[i % len(colors)], alpha=0.7)

plt.title("3D Bar Chart (Avoid)")
plt.show()
```

##  Best Practices for High-Quality Visualizations (with Code Examples)

###  Do: Use the Right Chart Type

| Question                   | Good Chart         |
| -------------------------- | ------------------ |
| How does one value change? | Line Plot          |
| Compare categories?        | Bar Chart          |
| Show distribution?         | Histogram, Boxplot |
| Show relationship?         | Scatterplot        |

```python
# Right Chart for Trend (Line Plot)
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y)
plt.title("Line Plot for Trend")
plt.show()

# Right Chart for Comparison (Bar Plot)
import seaborn as sns
categories = ["Cat A", "Cat B", "Cat C"]
values = [10, 20, 15]
sns.barplot(x=categories, y=values)
plt.title("Bar Plot for Categories")
plt.show()
```

###  Do: Use Preattentive Attributes

```python
# # Using color and size to highlight a key value
x = [1, 2, 3, 4, 5]
y = [5, 4, 3, 2, 10]
sizes = [40, 80, 100, 200, 300]
sns.scatterplot(x=x, y=y, size=sizes, sizes=(40, 200), hue=y, palette='coolwarm')
plt.title("Highlight with Preattentive Features")
plt.show()
```

###  Do: Keep it Simple

```python
# Minimalist chart
x = ["Q1", "Q2", "Q3", "Q4"]
y = [100, 110, 105, 120]
sns.barplot(x=x, y=y, color='skyblue')
plt.title("Quarterly Sales")
plt.xlabel("")
plt.ylabel("")
plt.show()
```
