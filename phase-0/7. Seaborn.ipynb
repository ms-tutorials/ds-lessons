## 1. What is Seaborn?

> **Story:** Imagine Matplotlib as your basic crayon set. It's great, but Seaborn is like upgrading to a full artist's toolkit with markers, brushes, and beautiful colors - all ready to use.

**Seaborn** is a powerful Python data visualization library built on top of Matplotlib. It makes it easier to create beautiful and informative charts with less code.


---

## 2.  Seaborn and Matplotlib

Seaborn uses Matplotlib under the hood but gives you more attractive and complex plots out-of-the-box.

### Example:

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = [1, 2, 2, 3, 3, 4, 4, 5]

# Matplotlib
plt.hist(data)
plt.title("Matplotlib Histogram")
plt.show()

# Seaborn
sns.histplot(data)
plt.title("Seaborn Histogram")
plt.show()
```

### Breakdown:

- `sns.histplot(data)` gives a styled histogram without manually setting grid, colors, etc.
- Seaborn integrates better with Pandas DataFrames.

## 3. When to Use Each

| Use Case              | Recommendation |
| --------------------- | -------------- |
| Fast, basic plotting  | Matplotlib     |
| Statistical plots     | Seaborn        |
| Plots with DataFrames | Seaborn        |
| Fine-tuned control    | Matplotlib     |

> Think of Matplotlib as the car engine, and Seaborn as the dashboard and interior - nicer and more comfortable, but built on the same system.

## 4. Basic Plots with Seaborn

Let's load a sample dataset and make basic plots.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")  # Tip amounts from a restaurant

sns.scatterplot(x="total_bill", y="tip", data=df)
plt.title("Scatter plot of Bill vs Tip")
plt.show()
```

### Breakdown:

- `load_dataset("tips")`: loads a built-in dataset.
- `scatterplot(...)`: shows relationship between two numeric variables.

---

## 5. Simple Univariate Boxplot

> Boxplots are like summaries in a box - they tell us the minimum, maximum, median, and outliers.

```python
sns.boxplot(x=df["total_bill"])
plt.title("Boxplot of Total Bill")
plt.show()
```

### What it shows:

- Line inside = median
- Box = interquartile range (middle 50%)
- Dots = outliers

---

## 6. Boxplot Grouped by Categorical Variable

```python
sns.boxplot(x="day", y="total_bill", data=df)
plt.title("Boxplot of Bills by Day")
plt.show()
```

### Breakdown:

- `x="day"`: groups data by categorical variable.
- `y="total_bill"`: numeric variable being measured.

---

## 7. Nested Boxplot with Grouping

Group by **day** and split by **sex**:

```python
sns.boxplot(x="day", y="total_bill", hue="sex", data=df)
plt.title("Bills by Day and Gender")
plt.show()
```

### Breakdown:

- `hue="sex"`: adds an inner split (color-coded) to each group.

---

## 8. Regression Plots

> Want to see if two variables move together in a straight-line relationship? Regression plots are perfect!

```python
sns.lmplot(x="total_bill", y="tip", data=df)
plt.title("Regression Plot of Tip vs Total Bill")
plt.show()
```

###  Breakdown:

- Adds a **trend line** and **confidence interval** to scatter plots.
